```python
## Pandas Input/Output Example
# This script demonstrates how to read data from a CSV file using the Pandas library.
# we use utilize the pd.read_csv function to read a CSV file

import pandas as pd
import numpy as np
import requests


## Fetch the data from the URL
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
response = requests.get(data_url)

with open ("imports-85.csv", "wb") as data_file:
    data_file.write(response.content)


# Define the column headers
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
##print("headers\n", headers)

# Read the CSV file into a DataFrame
df = pd.read_csv("imports-85.csv", names=headers)
##print(df.head(10))  # Display the first 10 rows of the DataFrame

# to save csv file
df.to_csv("automobile.csv", index=False, header= False)

# Get summary statistics of the DataFrame
print(df.describe())
#print(df.describe(include = "all"))
```
